# Assignment 1: Temperature Check 
Problem: 
Write a program that takes the current temperature as input and prints: 
● "It's too cold!" if below 10°C 
● "It's cool outside" if between 10°C and 25°C 
● "It's hot outside!" if above 25°C 

```python
temp = float(input("Enter current temperature in Celsius: "))

if temp < 10:
    print("It's too cold!")
elif 10 <= temp <= 25:
    print("It's cool outside") 
else:
    print("It's hot outside!")
```

# Assignment 2: ATM Withdrawal System 

Problem: 
Ask the user for their account balance and the amount they want to withdraw. 
● If the withdrawal amount is greater than the balance → print "Insufficient 
Balance" 
● If the withdrawal amount is less than or equal to balance → print "Transaction 
Successful" and show the remaining balance. 

```python
balance = float(input("Enter your account balance: "))
withdraw = float(input("Enter the amount you want to withdraw: "))

if withdraw > balance:
    print("Insufficient Balance")
else:
    balance -= withdraw
    print(f"Transaction Successful. Remaining balance: {balance}")
```

# Assignment 3: Student Grade Calculator

Problem:
Input marks (0–100) and print grade based on:
● 90–100 → A+
● 80–89 → A
● 70–79 → B
● 60–69 → C
● Below 60 → Fail

```python
marks = int(input("Enter marks (0-100): "))

if 90 <= marks <= 100:
    print("Grade: A+")
elif 80 <= marks <= 89:
    print("Grade: A")
elif 70 <= marks <= 79:
    print("Grade: B")
elif 60 <= marks <= 69:
    print("Grade: C")
else:
    print("Grade: Fail")
```

# Assignment 4: Bus Ticket Discount

Problem:
Ask the user for age and print ticket price:
● Below 5 → Free
● 5–18 → 50% Discount
● 60 or above → 30% Discount
● Others → Full Price

```python
age = int(input("Enter your age: "))

if age < 5:
    print("Ticket Price: Free")
elif 5 <= age <= 18:
    print("Ticket Price: 50% Discount")
elif age >= 60:
    print("Ticket Price: 30% Discount")
else:
    print("Ticket Price: Full Price")
```

# Assignment 5: Restaurant Bill with Discount

Problem:
Take total bill amount from the user:
● If the bill is above 1000 → apply 10% discount
● Otherwise → no discount
Then print the final bill amount.

```python
bill = float(input("Enter total bill amount: "))

if bill > 1000:
    bill = bill * 0.9  # Apply 10% discount
    print("10% discount applied.")
else:
    print("No discount applied.")

print(f"Final bill amount: {bill}")
```

# Assignment 7: Traffic Light System

Problem:
Ask the user to input a color (red, yellow, or green).
Print:
● "Stop" for red
● "Ready to go" for yellow
● "Go" for green
● "Invalid color" otherwise

```python
color = input("Enter traffic light color (red, yellow, or green): ").lower()

if color == "red":
    print("Stop")
elif color == "yellow":
    print("Ready to go")
elif color == "green":
    print("Go")
else:
    print("Invalid color")
```

# Assignment 8: Leap Year Checker

Problem:
Take a year as input and check:
● If it is divisible by 400 → Leap Year
● Else if divisible by 4 but not by 100 → Leap Year
● Otherwise → Not a Leap Year

```python
year = int(input("Enter a year: "))

if year % 400 == 0:
    print("Leap Year")
elif year % 4 == 0 and year % 100 != 0:
    print("Leap Year")
else:
    print("Not a Leap Year")
```

# Assignment 10: Electricity Bill Calculator

Problem:
Input number of units consumed and calculate bill:
● Up to 100 units → 5 BDT per unit
● 101–200 units → 7 BDT per unit
● Above 200 units → 10 BDT per unit
Print total payable amount.

```python
units = float(input("Enter number of units consumed: "))

if units <= 100:
    total_bill = units * 5
elif units <= 200:
    total_bill = units * 7
else:
    total_bill = units * 10

print(f"Total payable amount: {total_bill} BDT")
```
